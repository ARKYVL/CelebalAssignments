```python

import pandas as pd
import numpy as np


df = pd.read_csv('ecommerce_dataset.csv')


df['quantity'] = np.random.randint(1, 5, size=len(df))


print("Dataset loaded successfully!")
```

```python
print("Dataset Shape (Rows, Columns):", df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nLast 5 rows:")
display(df.tail()) 
```

```python

print("Missing values BEFORE cleaning:")
print(df.isnull().sum())


df.dropna(subset=['title', 'final_price'], inplace=True)


if 'rating' in df.columns:
    median_rating = df['rating'].median()
    df['rating'] = df['rating'].fillna(median_rating)


if 'product_description' in df.columns:
    df['product_description'] = df['product_description'].fillna("No description available")


print("\nMissing values AFTER cleaning:")
print(df.isnull().sum())
```

```python
highly_rated_df = df[df['rating'] > 4.5]

print(f"Total highly rated products: {len(highly_rated_df)}")
print("Preview of highly rated products:")
display(highly_rated_df.head(3)) 

subset_df = df[['title', 'final_price', 'quantity']]

print("\n Selected Columns Subset ")
display(subset_df.head(3))
```

```python

rows_before = len(df)
print(f"Rows before removing duplicates: {rows_before}")


df.drop_duplicates(inplace=True)


rows_after = len(df)
print(f"Rows after removing duplicates: {rows_after}")
print(f"Total duplicates removed: {rows_before - rows_after}")
```

```python
df['total_amount'] = df['final_price'] * df['quantity']

print("Preview of the new 'total_amount' column:")
display(df[['title', 'final_price', 'quantity', 'total_amount']].head()) 
```

```python
output_filename = 'final.csv'

df.to_csv(output_filename, index=False)

print(f"Success! Your clean dataset has been saved as: {output_filename}")
```
